## 2015년도에 평가된 모든 Romance 영화의 평균 평점은?

```python

from datetime import datetime
import time

start = '2015-01-01 0:0:1'
start_timestamp = time.mktime(datetime.strptime(start,'%Y-%m-%d %H:%M:%S').timetuple())

end = '2015-12-31 23:59:59'
end_timestamp = time.mktime(datetime.strptime(end,'%Y-%m-%d %H:%M:%S').timetuple())

# print(start_timestamp)
# print(end_timestamp)

date_mask = ((df_rating['timestamp'].values >= start_timestamp) & (df_rating['timestamp'].values <= end_timestamp))

result1 = pd.merge(df_rating[date_mask],df,
                   on='movieId', how='inner')

print(result1.loc[result1['genres'].str.contains('Romance')]['rating'].mean())
# result1.loc[result1['genres'].str.contains('Romance')]['rating']
```

```python
# 3. 연령대에 따른 월급 차이
# 30세 미만을 초년(young), 
# 30~59세 : 중년(middle), 
# 60세 이상 : 노년(old)
# 위의 범주로 연령대에 따른 월급의 차이를 알아보자

# 연령 별로 월급 평균을 구해서 디스플레이 하기
```

```python
r_df['age'] = 2020 - r_df['h10_g4'] + 1
y_mask = r_df['age'] < 30
m_mask = (r_df['age'] >= 30) & (r_df['age'] < 60)
o_mask = r_df['age'] >= 60

y_p = int(r_df.loc[y_mask,['p1002_8aq1'][0]].mean())
m_p = int(r_df.loc[m_mask,['p1002_8aq1'][0]].mean())
o_p = int(r_df.loc[o_mask,['p1002_8aq1'][0]].mean())

print("초년의 월급의 평균은 : {}만원 ".format(y_p))
print("중년의 월급의 평균은 : {}만원 ".format(m_p))
print("노년의 월급의 평균은 : {}만원 ".format(o_p))
```

```python
# 4. 연령대 및 성별 월급 차이
# 성별 월급 차이는 연령대에 따라 다른 양상을 보일 수 있습니다.
# 성별 월급 차이가 연령대에 따라 다른지 분석해보자

# 기존에는 3그룹(초년,중년,노년)이었지만 이젠 6그룹으로
# 그룹핑을 해야 한다.(초년남성,초년여성,..)
```

```python
ym_mask = (r_df['age'] < 30) & (r_df['h10_g3'] == 1)
ys_mask = (r_df['age'] < 30) & (r_df['h10_g3'] == 2)
mm_mask = ((r_df['age'] >= 30) & (r_df['age'] < 60)) & (r_df['h10_g3'] == 1)
ms_mask = ((r_df['age'] >= 30) & (r_df['age'] < 60)) & (r_df['h10_g3'] == 2)
om_mask = (r_df['age'] >= 60) & (r_df['h10_g3'] == 1)
os_mask = (r_df['age'] >= 60) & (r_df['h10_g3'] == 2)

ym_p = int(r_df.loc[ym_mask,['p1002_8aq1'][0]].mean())
ys_p = int(r_df.loc[ys_mask,['p1002_8aq1'][0]].mean())
mm_p = int(r_df.loc[mm_mask,['p1002_8aq1'][0]].mean())
ms_p = int(r_df.loc[ms_mask,['p1002_8aq1'][0]].mean())
om_p = int(r_df.loc[om_mask,['p1002_8aq1'][0]].mean())
os_p = int(r_df.loc[os_mask,['p1002_8aq1'][0]].mean())

print("초년남성의 월급의 평균은 : {}만원 ".format(ym_p))
print("초년여성의 월급의 평균은 : {}만원 ".format(ys_p))
print("중년남성의 월급의 평균은 : {}만원 ".format(mm_p))
print("중년여성의 월급의 평균은 : {}만원 ".format(ms_p))
print("노년남성의 월급의 평균은 : {}만원 ".format(om_p))
print("노년여성의 월급의 평균은 : {}만원 ".format(os_p))
```
