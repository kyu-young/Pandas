##  사용자가 평가한 모든 영화의 전체 평균 평점

```python
print('전체 영화 평점 평균')
print(rating_df['rating'].mean())
```

## 각 사용자별 평균 평점

```python
display(rating_df[['userId', 'rating']].groupby('userId').mean())
```

##  각 영화별 평균 평점

```python
movie_mean = rating_df[['movieId', 'rating']].groupby('movieId').mean()
movie_mean = pd.merge(movie_mean, movie_df[['movieId', 'title']], how='left', left_on=movie_mean.index, right_on='movieId')
display(movie_mean)
```

## 평균 평점이 가장 높은 영화의 제목(동률이 있을 경우 모두 출력)

```python
movie_mean[movie_mean['rating']==movie_mean['rating'].max()].sort_values('title')
```

##  Comedy영화 중 가장 평점이 낮은 영화의 제목

```python
movie_mean = pd.merge(movie_mean, movie_df[['movieId', 'genres']], how='left', on='movieId')
```

```python
comedy_movie = movie_mean[movie_mean['genres'].str.contains('Comedy')].reset_index(drop=True)
```

```python
display(comedy_movie[comedy_movie['rating']==comedy_movie['rating'].min()].sort_values('title'))
```

## 성별에 따른 월급 차이 과거에 비해 여성의 사회 진출이 활발하지만 직장에서의 위상에서는 여전히 차별이 존재하고 있는것이 사실. 실제로 그러한지 월급의 차이를 이용하여 사실을 확인해보자 

```python
df[['h10_g3', 'p1002_8aq1']].groupby('h10_g3').mean()
# 1이 남자 2가 여자
```

##  나이와 월급의 관계 몇 살 때 월급을 가장 많이 받을까? 또 그때의 월급은 얼마인가?

```python
df['age'] = df['h10_g4'].apply(lambda x: 2020-x+1)
```

```python
df[['age', 'p1002_8aq1']].groupby('age').mean().sort_values('p1002_8aq1', ascending=False).iloc[:1]
```

##  연령대에 따른 월급 차이 30세 미만을 초년(young), 30~59세 : 중년(middle), 60세 이상 : 노년(old) 위의 범주로 연령대에 따른 월급의 차이를 알아보자

```python
def age_group(age):
    if age < 30:
        return 'young'
    elif age <=59:
        return 'middle'
    else:
        return 'old'

df['age_group'] = df['age'].apply(lambda x: age_group(x))
```

```python
df[['age_group', 'p1002_8aq1']].groupby('age_group').mean()
```

## 연령대 및 성별 월급 차이 성별 월급 차이는 연령대에 따라 다른 양상을 보일 수 있습니다. 성별 월급 차이가 연령대에 따라 다른지 분석해보자 기존에는 3그룹(초년,중년,노년)이었지만 이젠 6그룹으로 그룹핑을 해야 한다.(초년남성,초년여성,..)

```python
df[['age_group', 'h10_g3', 'p1002_8aq1']].groupby(['age_group', 'h10_g3']).mean()
```
