```python
## Exercise - 1

import numpy as np
import pandas as pd

df_movie = pd.read_csv('./data/movie/movies.csv')
df_rating = pd.read_csv('./data/movie/ratings.csv')

print(df_movie.shape)  # (9742, 3)  영화정보
print(df_rating.shape)  # (100836, 4)  평점정보
```

```python
# 1. 사용자가 평가한 모든 영화의 전체 평균 평점
# display(df_rating.head())

# rating안의 결측치 포함여부 확인
# print(df_rating['rating'].isnull().sum())  # 0 => 결측치 없음

print(df_rating['rating'].mean())
```

```python
# 2. 각 사용자별 평균 평점

print(df_rating['rating'].groupby(df_rating['userId']).mean())
```

```python
# 3. 각 영화별 평균 평점

s1 = df_rating['rating'].groupby(df_rating['movieId']).mean()

result = pd.merge(s1, df_movie, left_index=True, right_on='movieId', how='inner')

display(result[['movieId', 'title', 'rating']])
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목(동률이 있을 경우 모두 출력)

s1 = df_rating['rating'].groupby(df_rating['movieId']).mean()

high_rating = s1[s1 == s1.max()]

result = pd.merge(high_rating, df_movie, left_index=True, right_on='movieId', how='inner')

display(result.sort_values(by='title')[['movieId','title','rating']])
```

```python
# 5. Comedy영화 중 가장 평균 평점이 낮은 영화의 제목

comedy_movie_id = df_movie.loc[df_movie['genres'].str.contains('Comedy')]['movieId']

comedy_mask = df_rating['movieId'].isin(comedy_movie_id)

low_rating = df_rating.loc[comedy_mask].groupby('movieId').mean()['rating'].min()

low_rating_mask = df_rating.loc[comedy_mask].groupby('movieId').mean()['rating'] == low_rating

low_movie_id = df_rating.loc[comedy_mask].groupby('movieId').mean().loc[low_rating_mask].index

low_movie_id = list(low_movie_id)

display(df_movie.loc[df_movie['movieId'].isin(low_movie_id)].sort_values(by='title'))
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 2015년도 평균 평점은?

from datetime import datetime
import time

start = '2015-01-01 0:0:1'
start_timestamp = time.mktime(datetime.strptime(start,'%Y-%m-%d %H:%M:%S').timetuple())

end = '2015-12-31 23:59:59'
end_timestamp = time.mktime(datetime.strptime(end,'%Y-%m-%d %H:%M:%S').timetuple())

# print(start_timestamp)
# print(end_timestamp)

date_mask = ((df_rating['timestamp'].values >= start_timestamp) & (df_rating['timestamp'].values <= end_timestamp))

result1 = pd.merge(df_rating[date_mask],df_movie,
                   on='movieId', how='inner')

print(result1.loc[result1['genres'].str.contains('Romance')]['rating'].mean())
# result1.loc[result1['genres'].str.contains('Romance')]['rating']
```
